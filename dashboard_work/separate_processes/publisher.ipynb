```python
import pandas as pd
import numpy as np
from streamz import Stream
from dask.distributed import Client, Pub, Sub
import warnings
import time
warnings.filterwarnings('ignore')

# client = Client(n_workers=8)
client = Client(n_workers=1)
pub = Pub('test',client=client)
print(client.scheduler.address)
client
```

```python
def _put(x):
    pub.put(x)

# Some pipeline to be defined
source = Stream()
source.sink(_put); # need semi colon to avoid display error
```

```python
for i in range(30):
    # Sending random dataframes with x,y,z
    x = np.random.rand(1)
    y = np.random.rand(1)
    z = np.random.rand(1)
    time.sleep(1)
    source.emit(pd.DataFrame({"x":x,"y":y,"z":z}).set_index("x"))
```
